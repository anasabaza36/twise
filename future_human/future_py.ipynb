```python
import cv2
import numpy as np
import mediapipe as mp
from deepface import DeepFace
from PIL import Image
import matplotlib.pyplot as plt
import tensorflow as tf # explicitly import tensorflow

# Charger une image (exemple d'un fichier sur Colab)
from google.colab import files

uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Lire l'image
image = Image.open(image_path)
image_np = np.array(image)

# Convertir en BGR pour OpenCV
image_np = cv2.cvtColor(image_np, cv2.COLOR_RGB2BGR)

# Détection du visage
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils

with mp_face_detection.FaceDetection(min_detection_confidence=0.5) as face_detection:
    results = face_detection.process(cv2.cvtColor(image_np, cv2.COLOR_BGR2RGB))

    if results.detections:
        for detection in results.detections:
            mp_drawing.draw_detection(image_np, detection)

# Afficher l'image avec les visages détectés
plt.imshow(cv2.cvtColor(image_np, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

# Analyse du visage avec DeepFace
# Use the image path instead of the numpy array
analysis = DeepFace.analyze(img_path = image_path, actions=['age', 'gender'], enforce_detection=False)
age = analysis[0]['age']
gender = analysis[0]['dominant_gender']

print(f"Âge estimé: {age}")
print(f"Genre: {gender}")
```

```python
from fastapi import FastAPI, UploadFile, File
from io import BytesIO
from PIL import Image
import numpy as np
import cv2
from deepface import DeepFace
import nest_asyncio # Import nest_asyncio

app = FastAPI()  # ✅ Déclare l'application FastAPI ici

# Base de styles par domaine scientifique
style_database = {
    "astronomie": {
        "homme": ["Veste en cuir + chemise sombre", "Pantalon noir", "Lunettes futuristes"],
        "femme": ["Blazer bleu foncé", "Jean slim", "Bottes montantes"]
    },
    "biotechnologie": {
        "homme": ["Blouse blanche + T-shirt scientifique", "Jean classique", "Sneakers"],
        "femme": ["Blouse blanche", "Pantalon ajusté", "Lunettes de protection"]
    },
    "intelligence artificielle": {
        "homme": ["Hoodie tech noir", "Jean slim", "Baskets modernes"],
        "femme": ["Blazer minimaliste", "Pantalon slim", "Chaussures futuristes"]
    }
}

def get_outfit_recommendation(domain, gender):
    gender = "homme" if gender == "Man" else "femme"

    if domain in style_database:
        return style_database[domain].get(gender, ["Style non défini"])
    else:
        return ["Domaine non reconnu, essayez un autre."]

@app.post("/recommend-style/")
async def recommend_style(file: UploadFile = File(...), domain: str = "astronomie"):
    try:
        # Lire et traiter l'image
        image = Image.open(BytesIO(await file.read()))
        image_np = np.array(image)
        image_rgb = cv2.cvtColor(image_np, cv2.COLOR_BGR2RGB)

        # Analyse du visage avec DeepFace
        analysis = DeepFace.analyze(image_rgb, actions=['age', 'gender'], enforce_detection=False)

        age = analysis[0]['age']
        gender = analysis[0]['dominant_gender']

        # Récupérer la recommandation de style
        outfit = get_outfit_recommendation(domain, gender)

        return {
            "age": age,
            "gender": gender,
            "domain": domain,
            "recommended_outfit": outfit
        }

    except Exception as e:
        return {"error": str(e)}

if __name__ == "__main__":
    import uvicorn
    nest_asyncio.apply() # Apply nest_asyncio to patch the event loop
    uvicorn.run(app, host="0.0.0.0", port=8000)
```

```python
from fastapi import FastAPI, File, UploadFile, Form, HTTPException
import uvicorn
import cv2
import numpy as np
from PIL import Image
import io
import nest_asyncio

app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "🚀 FastAPI server is running! Use POST /analyze to send an image."}

@app.post("/analyze")
async def analyze_image(file: UploadFile = File(...), domain: str = Form(...)):
    try:
        if file is None:
            raise HTTPException(status_code=400, detail="No file received")

        image_bytes = await file.read()
        if not image_bytes:
            raise HTTPException(status_code=400, detail="Empty file received")

        image = Image.open(io.BytesIO(image_bytes))
        image_np = np.array(image)

        # Convert to grayscale
        gray_image = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)

        # Resize image
        resized_image = cv2.resize(gray_image, (224, 224))

        # Normalize pixel values
        normalized_image = resized_image / 255.0

        return {"domain": domain, "message": "✅ Image processed successfully"}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

if __name__ == "__main__":
   nest_asyncio.apply()
    uvicorn.run(app, host="0.0.0.0", port=8000)
```
